```python
# SQL Alchemy
#import dependencies
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:Mfon@Learn2019@localhost:5432/SQL-Homework')
connection = engine.connect()

# Pandas
import pandas as pd

# Matplotlib
import matplotlib.pyplot as plt

# NumPy
import numpy as np
```

```python
# Select data in the Salaries Table
salary_data = pd.read_sql("SELECT * FROM salaries", connection)
salary_data.head()
```

```python
# Select data from the Titles Table
titles_data = pd.read_sql("SELECT * FROM titles", connection)
titles_data.head()
```

```python
# Merge Columns and Combine Salary and Title Data
combined_data = pd.merge(salary_data, titles_data, on="emp_no", how="inner")
combined_data.head()
```

```python
# Group the data by Title and find the average salary
title_group_df = combined_data.groupby("title").mean()
title_group_df
```

```python
# Show Only title & salary data for the bar chart
title_Sal_df = title_group_df.drop('emp_no', axis=1)
title_Sal_df
```

```python
# Reset Index
title_avg_sal_df = title_Sal_df.reset_index()
title_avg_sal_df
```

```python
# Set x_axis, y_axis & Tick Locations
x_axis = title_avg_sal_df["title"]
ticks = np.arange(len(x_axis))
y_axis = title_avg_sal_df["salary"]
 
# Create Bar Chart Based on Data
plt.bar(x_axis, y_axis, align="center", alpha=0.5, color=["k", "r", "g", "m", "b", "c", "y"])

# Create Ticks for Bar Chart's x_axis
plt.xticks(ticks, x_axis, rotation="vertical")

# Set Labels & Title
plt.ylabel("Salaries ($)")
plt.xlabel("Employee Titles")
plt.title("Average Employee Salary by Title")

# Save Figure
plt.savefig("Images/average_salary_by_title.png")

# Show plot
plt.show()
```
